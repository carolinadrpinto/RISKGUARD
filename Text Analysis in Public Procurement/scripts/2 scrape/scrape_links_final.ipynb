## Extrair os links


Pipeline para extrair link das peças do procedimento e link do anúncio - neste momento só concursos públicos 2023

Depois consultas prévias

Depois Ajuste Direto Regime geral

```python
data = []
with open('../data/contracts_links_OT.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        data.append(json.loads(line))
contracts_scraped = pd.DataFrame(data)
```

```python
contratos_scr = contracts_scraped['contract_id'].to_list()
ids = contratos[(contratos['Data Celebração'].dt.year==2023)  & (contratos['Tipo de procedimento']=='Concurso público') & (~(contratos['N.º Contrato'].isin(contratos_scr)))]['N.º Contrato'].head(400).to_list()
```

```python
len(ids)
```

```python
results = fetch_contract_links_in_batches(ids)
```

```python
out_path = Path("../data/contracts_links_OT.jsonl")

# write (overwrite). use mode="a" to append
with out_path.open("a", encoding="utf-8") as f:
    for rec in results:
        json.dump(rec, f, ensure_ascii=False)
        f.write("\n")
```
